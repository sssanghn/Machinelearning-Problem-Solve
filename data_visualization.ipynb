# 수치형 데이터 시각화

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

**히스토그램(histplot)**
> 히스토그램은 수치형 데이터의 구간별 빈도수를 나타내는 그래프이다. </br>
bins 파라미터 => 구간 개수 지정 </br>
hue 파라미터 => 빈도를 특정 범주 별로 구분 </br>
multiple='stack' 매개변수 => 누적 표현

```python
sns.histplot(data=titanic, x='age')
```

```python
sns.histplot(data=titanic, x='age', bins=10)
```

```python
sns.histplot(data=titanic, x='age', hue='alive', multiple='stack')
```

**커널밀도추정 함수 그래프(kdeplot)**
> 히스토그램을 매끄럽게 곡선으로 연결한 그래프

이산형 => 히스토그램 </br>
연속형 => 커널밀도추정 함수 그래프

```python
sns.kdeplot(data=titanic, x='age')
```

```python
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack')
```

**분포도(displot)**
> 수치형 데이터 하나의 분포를 나타내는 그래프 </br>
\* 파라미터만 조정하면 hisplot()과 kdeplot()이 제공하는 기본 그래프를 모두 그릴 수 있다.

```python
sns.displot(data=titanic, x='age')
```

```python
sns.displot(data=titanic, x='age', kind='kde')
```

> 히스토그램과 커널밀도추정 함수 그래프를 동시에 그릴 수도 있다. </br>
* kde=True

```python
sns.displot(data=titanic, x='age', kde=True)
```

**러그플롯(rugplot)**
> 주변 분포를 나타내는 그래프 </br>
단독으로 사용하기보다는 주로 다른 분포도 그래프와 함께 사용

```python
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age')
```

# 범주형 데이터 시각화

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

**막대 그래프(barplot)**
> 막대 그래프는 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 그린다. </br>
평균은 막대 높이로, 신뢰구간은 오차 막대로 표현한다. </br>
barplot()은 원본 데이터 평균이 아니라 샘플링한 데이터의 평균과 신뢰구간을 구한다.

```python
sns.barplot(x='class', y='fare', data=titanic)
```

**포인트플롯(pointplot)**
> 막대 그래프와 모양만 다를 뿐 동일한 정보를 제공한다. </br>
그래프를 점과 선으로 나타낸다.

> "언제 포인트플롯을 사용하는게 좋을까?" <br/>
* 한 화면에 여러 그래프를 그릴 때이다.

```python
sns.pointplot(x='class', y='fare', data=titanic)
```

**박스플롯(boxplot)**
> 박스플롯은 막대 그래프나 포인트플롯보다 더 많은 정보를, 구체적으로 5가지 요약 수치를 제공한다. </br>
5가지 요약 수치는 최솟값, 제1사분위 수, 제2사분위 수, 제3사분위 수, 최댓값을 뜻한다. </br>
사분위 범위 수(IQR): Q3 - Q1 </br>
최댓값: Q3 + 1.5 * IQR </br>
최솟값: Q1 - 1.5 * IQR </br>

```python
sns.boxplot(x='class', y='age', data=titanic)
```

**바이올린플롯(violinplot)**
> 박스플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프이다. </br>
* 각 범주별로 5가지 요약 수치를 한눈에 보고 싶다면 박스플롯
* 수치형 데이터의 전체적인 분포 양상을 알고 싶다면 바이올린플롯

> split=True를 전달하면 hue에 전달한 피처를 반으로 나누어 보여준다.

```python
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True)
```

**카운트플롯(countplot)**
> 범주형 데이터의 개수를 확인할 때 사용한다.

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(y='class', data=titanic)
```

**barplot() vs countplot()**
> barplot(): 범주형 데이터 별 수치형 데이터의 평균(피처를 두 개 받음) </br>
* <mark>여기서 평균은 샘플링한 데이터의 평균이다</mark>

> countplot(): 범주형 데이터의 개수(피처를 한 개 받음)

**파이 그래프(pie)**
> 파이 그래프는 범주형 데이터별 비율을 알아볼 때 사용하기 좋은 그래프이다. </br>
seaborn에서 파이 그래프를 지원하지 않아 matplotlib을 사용해야 한다. </br>
* autopct 파라미터를 통해 비율을 숫자로 나타낼 수 있다.

```python
import matplotlib.pyplot as plt

x = [10, 60, 30]
labels = ['A', 'B', 'C']

plt.pie(x=x, labels=labels, autopct='%.1f%%')
```

# 데이터 관계 시각화

> 관계도는 여러 데이터 사이의 관계를 살펴보기 위한 그래프이다. </br>
히트맵, 라인플롯, 산점도, 회귀선을 포함한 산점도가 있다.

**히트맵(heatmap)**
> 데이터간 관계를 색상으로 표현한 그래프이다.

```python
import seaborn as sns

flights = sns.load_dataset('flights')
```

```python
flights.head()
```

> 히트맵을 그리는 데 활용하려면 데이터 구조를 조금 바꿔야 한다. </br>
<mark>판다스의 pivot() 함수를 활용하자.</mark> </br>
pivot() 함수는 index와 columns 파라미터에 전달한 피처를 각각 행과 열로 지정하고, values 파라미터에 전달한 피처를 합한 표를 반환한다.

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

**라인플롯(lineplot)**
> 두 수치형 데이터 사이의 관계를 나타낼 때 사용한다. </br>
* x 파라미터에 전달한 값에 따라 y 파라미터에 전달한 값의 평균과 95% 신뢰구간을 나타낸다.

```python
sns.lineplot(x= 'year', y='passengers', data=flights)
```

**산점도(scatterplot)**
> 두 데이터의 관계를 점으로 표현


```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

**회귀선을 포함한 산점도 그래프(regplot)**
> regplot()은 산점도와 선형 회귀선을 동시에 그려주는 함수이다.

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

> 신뢰구간을 99%로 늘리려면 ci 파라미터에 99를 전달하면 된다.

```python
sns.regplot(x='total_bill', y='tip', ci=99, data=tips)
```
